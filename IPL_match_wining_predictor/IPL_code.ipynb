We just need the data of the first innings beacuse by knowing the first innings score we can predict which team can win

```python
total_score_df = total_score_df[total_score_df["inning"] == 1] 
total_score_df                           ##Here we r extracting the only second innings score...
```

```python
match_df = match.merge(total_score_df[["match_id","total_runs"]],left_on="id",right_on="match_id")
match_df              ##Here that dataframe has been merge to the match dataframe...
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
               ##Here vr replacing the names of the teams...
```

```python
match_df = match_df[match_df["team1"].isin(teams)]
match_df = match_df[match_df["team2"].isin(teams)]
```

```python
match_df.shape                  ##See the before shape n the present shape.. it will be different 
```

```python
match_df["dl_applied"].value_counts()       ##There r 15 matchs affected by the rain...i dont want that matchs
```

```python
match_df = match_df[match_df["dl_applied"] == 0]
match_df                       ##This is the dataframe of the matches that didn't affect by the rain...
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
match_df
```

```python
delivery_df = match_df.merge(delivery,on="match_id")
delivery_df
         ##In this data we have the data of the first n seconds both the innings scores....but we need only second innings score so we need to remove the first innings score...
```

```python
delivery_df = delivery_df[delivery_df["inning"] == 2]
delivery_df
```

```python
delivery_df['current_score'] = delivery_df.groupby("match_id")['total_runs_y'].cumsum()
```

```python
delivery_df         ##Here the new column will be added...
```

```python
delivery_df["runs_left"] = delivery_df["total_runs_y"] - delivery_df["current_score"]
```

```python
delivery_df                 ##New column is added...
```

```python
delivery_df["balls_left"] = 126 - (delivery_df["over"] * 6 + delivery_df["ball"])
                              ##See the code properly n apply the caluculations u will understand the code....
```

```python
delivery_df              ##Another column is added...
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum().values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
##crr = runs/overs
delivery_df["crr"] = (delivery_df["current_score"]*6)/(120 - delivery_df["balls_left"])
```

```python
# Replace negative values in 'runs_left' with 0
delivery_df['runs_left'] = delivery_df['runs_left'].apply(lambda x: max(x, 0))

# Calculate required run rate again
delivery_df['rrr'] = (delivery_df['runs_left'] / delivery_df['balls_left']) * 6
```

```python
delivery_df
```

```python
def result(row):
    return 1 if row["batting_team"] == row["winner"] else 0
```

```python
delivery_df["result"] = delivery_df.apply(result,axis=1)
```
